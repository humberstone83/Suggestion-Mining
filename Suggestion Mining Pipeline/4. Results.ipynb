```python
import pickle
import pandas as pd
pd.set_option('max_colwidth',40)
pd.set_option('display.max_rows',833)

import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

# Results
- This notebook is for viewing the results without having to first run the experiments. We also use graphs to visualize our results.

```python
def import_tfidf_results():
    Results_A = pickle.load( open( "results_A.pkl", "rb" ) )
    Results_B = pickle.load( open( "results_B.pkl", "rb" ) )
    return Results_A, Results_B
    
def import_nn_results():
    Results_A = pickle.load( open( "results_1.pkl", "rb" ) )
    Results_B = pickle.load( open( "results_2.pkl", "rb" ) )
    return Results_A, Results_B
    
def import_pp_results():
    Results_A = pickle.load( open( "results_3.pkl", "rb" ) )
    Results_B = pickle.load( open( "results_4.pkl", "rb" ) )
    return Results_A, Results_B   

def import_aug_results():
    Results_A = pickle.load( open( "results_5.pkl", "rb" ) )
    Results_B = pickle.load( open( "results_6.pkl", "rb" ) )
    return Results_A, Results_B   

def import_pos_results():
    Results_A = pickle.load( open( "results_7.pkl", "rb" ) )
    Results_B = pickle.load( open( "results_8.pkl", "rb" ) )
    return Results_A, Results_B    
```

```python
def plot_results(results):
    accuracy = results['Accuracy'].tolist()
    precision = results['Precision'].tolist()
    recall = results['Recall'].tolist()
    fscore =  results['F-Score'].tolist()
    index = results['Classifier'].tolist()

    df = pd.DataFrame({'Accuracy': accuracy,
                   'Precision': precision, 'Recall' : recall, 'F-Score' : fscore}, index=index)
    df.plot.barh()
```

# TD-IDF Experiment:

```python
print("TD-IDF Experiment:")
print()
results_A, results_B = import_tfidf_results()
print("Dataset A:")
print(results_A)

plot_results(results_A)
print()
print("Dataset B:")
print(results_B)
print()
plot_results(results_B)
```

# Pre-processing Techniques Experiment:

```python
print("Pre-processing Experiment:")
print()
results_A, results_B = import_nn_results()
print("Dataset A:")
print("Basic Clean applied only...")
print(results_A)
plot_results(results_A)
print()
print("Stop words removed & lemmatized...")
results_A2, results_B2 = import_pp_results()
print(results_A2)
plot_results(results_A2)
print()
print("Dataset B:")
print("Basic Clean applied only...")
print(results_B)
print()
plot_results(results_B)
print("Stop words removed & lemmatized...")
print(results_B2)
print()
plot_results(results_B2)
```

# Data Augmentation Experiment:

```python
print("Data Augmentation Experiment:")
print()
results_A, results_B = import_nn_results()
print("Dataset A:")
print("Basic Clean applied only...")
print(results_A)
plot_results(results_A)
print()
print("Augmented Training Data...")
results_A2, results_B2 = import_aug_results()
print(results_A2)
plot_results(results_A2)
print()
print("Dataset B:")
print("Basic Clean applied only...")
print(results_B)
print()
plot_results(results_B)
print("Augmented Training Data...")
print(results_B2)
print()
plot_results(results_B2)
```

```python
print("Part of Speech Tags Experiment:")
print()
results_A, results_B = import_nn_results()
print("Dataset A:")
print("Basic Clean applied only...")
print(results_A)
plot_results(results_A)
print()
print("Part of Speech Tags Data...")
results_A2, results_B2 = import_pos_results()
print(results_A2)
plot_results(results_A2)
print()
print("Dataset B:")
print("Basic Clean applied only...")
print(results_B)
print()
plot_results(results_B)
print("Part of Speech Tags Data...")
print(results_B2)
print()
plot_results(results_B2)
```
